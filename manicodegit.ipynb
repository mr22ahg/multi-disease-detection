```python
def DataClnHrt(data_1): 
    ftct=data_1.dtypes[data_1.dtypes=='object'].index.tolist()   
    clean_fets=[[] for i in range(len(ftct))]
    for i in range(len(ftct)):
        for j in range(len(data_1[ftct[i]].tolist())): 
            if data_1[ftct[i]].tolist()[j]=="?":   
                clean_fets[i].append(0)    
            else:
                clean_fets[i].append(data_1[ftct[i]].tolist()[j]) 
        data_1[ftct[i]]=numpy.array(clean_fets[i],float) 
    print(data_1.info())
    return data_1 
```

```python
def Clean_Kidney_Data(data_2): 
    try:
        data_2['classification']=data_2['classification'].replace(data_2['classification'].unique(),['ckd','ckd','notckd'])
    except: 
        pass
    obdata_2=data_2.dtypes[data_2.dtypes=='object'].index.tolist() 
    nmdata_2=data_2.dtypes[data_2.dtypes!='object'].index.tolist()
    for oc in obdata_2:
        data_2[oc]=data_2[oc].fillna(data_2[oc].mode()[0])
    for nc in nmdata_2:
        data_2[nc]=data_2[nc].fillna(data_2[nc].mode()[0])
    transfets=['pcv', 'wc', 'rc']
    for t in transfets: 
        dtarr=numpy.array(data_2[t]) 
        formed_fets=[] 
        for x in dtarr:
            if x.isdigit()==False:
                formed_fets.append(0)
            else:
                formed_fets.append(x)
        data_2[t]=numpy.array(formed_fets,float)
    print(data_2.isna().sum())
    print(data_2.info())
    print(data_2['classification'].value_counts())
    return data_2
Data_2Kcln=Clean_Kidney_Data(Data_2K)
Data_2Kcln.head(10)
```

```python
def EncodingFunc(data, dn): 
    data_opr=data.copy()
    print("Information Before Encdoing {} Data:\n".format(dn))   
    print(data_opr.info()) 
    ctdata_opr=data_opr.dtypes[data_opr.dtypes=='object'].index.tolist()[:-1] 
    print(ctdata_opr)
    for c in ctdata_opr:
        data_opr[c]=data_opr[c].replace(data_opr[c].unique(),[x for x in range(len(data_opr[c].unique()))]) 
    print("Information After Encdoing {} Data:\n".format(dn))   
    print(data_opr.info())
    return data_opr
```

```python
alldt=[Data_1H, Data_2K]
HKDt=['Heart Disease','Kidney Disease'] 
EncDts=[]
for d in range(len(alldt)):
    EncDts.append(EncodingFunc(alldt[d], HKDt[d]))
EncDts[0].head()
```

```python
def OutRemFunc(dthk,xax,yax,col,tptx,iddt):
    arrdthk=numpy.array(dthk.iloc[:,:-1])   
    pcdthk = decomposition.PCA(n_components=2)
    pcdthk.fit(arrdthk)
    st1="Comp"
    st2="Variance"
    st3="Maximum Variance Value"
    vrls=pcdthk.explained_variance_ratio_
    pcdthkcm=["{}-{}".format(st1,i+1) for i in range(len(vrls.tolist()))]
    dsvs.figure(figsize=(xax,yax)) 
    dsvs.title("{}(PCA={})\n{} Normalization {} Data\n{}: {}".format(st2,2,tptx,iddt,st3,round(max(vrls),8)),fontsize=18)
    dsvs.bar(pcdthkcm,pcdthk.explained_variance_ratio_.tolist(),width=0.5,color=col)
    dsvs.xlabel("Components",fontsize=14)
    dsvs.ylabel("Variance",fontsize=14)
    dsvs.grid()
    dsvs.show()
    return vrls
```

```python
def SclFunc(df):
    prepproc = preprocessing.MinMaxScaler()
    nrm_df=prepproc.fit_transform(df)
    return nrm_df 
```

```python
Data_OutRem=[]
ix=0
for dtds in EncDts:
    vrrec=[] 
    vrrec.append(OutRemFunc(dtds.drop(dtds.columns.tolist()[-1],axis=1),6,3,"r","Before",HKDt[ix]))

    lst=[]  
    lnv=[]   
    for rp in vrrec: 
        for r in rp:
            if r>0.6:
                lnv.append(True)
    if len(lnv)==1 and True in lnv: 
        SclData=SclFunc(dtds.drop(dtds.columns.tolist()[-1],axis=1))
    SclData=pandas.DataFrame(SclData,columns=dtds.drop(dtds.columns.tolist()[-1],axis=1).columns.tolist())
    SclData[dtds.columns.tolist()[-1]]=dtds[dtds.columns.tolist()[-1]]
    # check outlier again
    OutRemFunc(SclData.drop(SclData.columns.tolist()[-1],axis=1),6,3,"g","After",HKDt[ix])
    Data_OutRem.append(SclData) 
    ix+=1
Data_OutRem[0].head(10)
```

```python
def FetModel1(dthk, iddt): 
    print("--------------------------------------------------------------")
    print("Feature Selection for {} Data using Chi-Squred".format(iddt))
    print("--------------------------------------------------------------")
    Xdthk=dthk.drop([dthk.columns.tolist()[-1]],axis=1)
    Ydthk=dthk[dthk.columns.tolist()[-1]] 
    Ydthk=Ydthk.replace(Ydthk.unique(),[x for x in range(len(Ydthk.unique()))]) 
    ftm1=feature_selection.SelectKBest(feature_selection.chi2, k=(len(dthk.columns)-1)-3)
    ftm1.fit(Xdthk, Ydthk)   # train model
    print("Selected Features for {} Data using Chi-Squred: \n".format(iddt),*Xdthk.columns[ftm1.get_support()],sep="\n")
    print("--------------------------------------------------------------")
    return Xdthk.columns[ftm1.get_support()]
```

```python
def FetModel2(dthk, iddt): 
    print("--------------------------------------------------------------")
    print("Feature Selection for {} Data using RFE".format(iddt))
    print("--------------------------------------------------------------")
    Xdthk=dthk.drop([dthk.columns.tolist()[-1]],axis=1)
    Ydthk=dthk[dthk.columns.tolist()[-1]] 
    Ydthk=Ydthk.replace(Ydthk.unique(),[x for x in range(len(Ydthk.unique()))])
    ftm1 = feature_selection.RFE(estimator=linear_model.LogisticRegression(),n_features_to_select = (len(dthk.columns)-1)-3, step = 0.7)
    ftm1fitted=ftm1.fit(Xdthk,Ydthk)   # train model
    print("Selected Features for {} Data using RFE: \n".format(iddt),*Xdthk.columns[ftm1fitted.get_support()],sep="\n")
    print("--------------------------------------------------------------")
    return Xdthk.columns[ftm1fitted.get_support()]
```

```python
SelecetdFets=[[],[]] 
for dn in range(len(Data_OutRem)):
    atrch=FetModel1(Data_OutRem[dn],HKDt[dn]) 
    atrrf=FetModel2(Data_OutRem[dn],HKDt[dn]) 
    for x in atrch:  
        if x in atrrf: 
            SelecetdFets[dn].append(x) 
```

```python
def DataPrepFunc(dta,selfn,nm): 
    datapr=dta[selfn]   
    datatrg=dta[dta.columns.tolist()[-1]] 
    display(HTML(datapr.head(10).to_html())) 
    trn_X, tst_X, trn_Y, tst_Y=model_selection.train_test_split(datapr, datatrg, test_size=0.25, random_state=42) # split data with                                                  
    return trn_X, tst_X, trn_Y, tst_Y
```

```python
XTrHK, XTsHK, yTrHK, yTsHK=[[],[],[],[]]     
for ndt in range(len(Data_OutRem)):   
    DataPreped=DataPrepFunc(Data_OutRem[ndt],SelecetdFets[ndt],HKDt[ndt])
    XTrHK.append(DataPreped[0])  
    XTsHK.append(DataPreped[1]) 
    yTrHK.append(DataPreped[2]) 
    yTsHK.append(DataPreped[3])
```

```python
def Prediction(dsdfs,clfapl,iddt,cmps):   
    MetHKRes=[]  
    for dsl in range(len(dsdfs)): 
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("              Executing {} Data for Disease Detection                        ".format(HKDt[dsl]))
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        MetHKResults=[
            [],
            [],
            [],
            [],
            [],
            []
        ]  
        print("_____________________________________________________________________________")
        for i in range(len(clfapl)):  
            print("                            {} ".format(iddt[i]))
            print("_____________________________________________________________________________")
            MetHKRes_loop=[
                [],
                [],
                [],
                [],
                [],
                []
            ]    
            for ts in range(10):  
                clfapl[i].fit(XTrHK[dsl],yTrHK[dsl]) 
                test_Predicted=clfapl[i].predict(XTsHK[dsl])   
                train_predicted=clfapl[i].predict(XTrHK[dsl]) 
                MetHKRes_loop[0].append(round(metrics.accuracy_score(yTsHK[dsl],test_Predicted)*100,4)) 

                evmt=metrics.precision_recall_fscore_support(yTsHK[dsl],test_Predicted, average='weighted')[:-1]
                a,b,c=round(evmt[0],4)*100,round(evmt[1],4)*100,round(evmt[2],4)*100,
                MetHKRes_loop[1].append(a)
                MetHKRes_loop[2].append(b)
                MetHKRes_loop[3].append(c)
                cmt=pandas.crosstab(yTsHK[dsl], test_Predicted, rownames=['Actual'], colnames=['Detected'], margins=True)
                MetHKRes_loop[4].append(cmt.iloc[:2,:2]) 
                MetHKRes_loop[5].append(metrics.classification_report(yTsHK[dsl], test_Predicted))   
            slrs=MetHKRes_loop[0].index(max(MetHKRes_loop[0]))   
            MetHKResults[0].append(round(metrics.accuracy_score(yTrHK[dsl],train_predicted)*100,4))   
            MetHKResults[1].append(MetHKRes_loop[0][slrs])
            MetHKResults[2].append(MetHKRes_loop[1][slrs]) 
            MetHKResults[3].append(MetHKRes_loop[2][slrs]) 
            MetHKResults[4].append(MetHKRes_loop[3][slrs]) 
            ovf_fitting=round(abs(round(metrics.accuracy_score(yTrHK[dsl],train_predicted)*100,2)-MetHKRes_loop[0][slrs]),2)
            MetHKResults[5].append(ovf_fitting)
            dsvs.figure(figsize=(5,3))
            dsvs.title("{} ({})".format(iddt[i],HKDt[dsl].split()[0]), fontsize=20,color="b")
            plotse.heatmap(MetHKRes_loop[4][slrs],fmt="d",annot=True,cmap=cmps[dsl])
            dsvs.show()
            print("\nTest Accuracy: {}%\n".format(MetHKRes_loop[0][slrs]))
            print("Train Accuracy: {}%\n".format(round(metrics.accuracy_score(yTrHK[dsl],train_predicted)*100,2)))
            print("\nClassification Report for {} \n\n{}".format(iddt[i],MetHKRes_loop[5][slrs]))
            print("_____________________________________________________________________________")

        DfResult=pandas.DataFrame({
            "Model":iddt,
            "Train_Accuracy({})".format(HKDt[dsl]):MetHKResults[0],
            "Test_Accuracy({})".format(HKDt[dsl]):MetHKResults[1],
            "Test_Precision({})".format(HKDt[dsl]):MetHKResults[2],
            "Test_Recall({})".format(HKDt[dsl]):MetHKResults[3],
            "Test_F1-Score({})".format(HKDt[dsl]):MetHKResults[4],
            "Overfit Model({})".format(HKDt[dsl]):MetHKResults[5]
        })
        Clrs=["#57FEFF","#FFFF00","#3EA99F","#EB5406","#F660AB"]
        for ar in DfResult.columns.tolist()[1:]:  
            dt_tmp=DfResult.sort_values(by=ar,ascending=True)
            dsvs.figure(figsize=(6,4))
            dsvs.title("Comparison of {}".format(ar),fontsize=18,color="#342D7E")
            dsvs.barh(dt_tmp["Model"],dt_tmp[ar],color=Clrs)
            dsvs.ylabel("Model",fontsize=16,color="#342D7E")
            dsvs.xlabel("{}".format(ar),fontsize=16,color="#342D7E")
            for slx, vl in enumerate(dt_tmp["{}".format(ar)]):
                dsvs.text(vl, slx, str(vl))
            dsvs.grid()
            dsvs.show()
        DfResult1=DfResult.sort_values(by="Test_Accuracy({})".format(HKDt[dsl]),ascending=False)
        DfResult1.to_csv("{}_Disease_Result.csv".format(HKDt[dsl]))
        MetHKRes.append(DfResult1.reset_index(drop=True))
        display(HTML(DfResult1.to_html()))
    return MetHKRes
```
